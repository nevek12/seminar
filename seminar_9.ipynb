5.5) Реализовать функции для выполнения четырех арифмитических операций преобразующих целые числа в целое число. Для двух заранее заданных целых чисел выполнить выбранную пользователем арифметическую операцию.

```python
def pro(k):
    if k == '*':
        return 10 * 5  
    elif k == '-':
        return 10 - 5
    elif k == '+':
        return 10 + 5
    elif k == '/':
        return 10 // 5
print(pro(input()))
```

5.6) Написать калькулятор для строковых выражений вида '<число> <операция> <число>', где <число> - целое число, например 113, <арифмитическая операция> - одна из операций +,-,*,/(деление нацело),%(остаток от деления),^(возведение в степень). Пример calc('13 - 5') -> 8

```python
def calc(op):
    if op.find('-') != -1:
        return int(op.split(' - ')[0]) - int(op.split(' - ')[1])
    elif op.find('*') != -1:
        return int(op.split(' * ')[0]) * int(op.split(' * ')[1])
    elif op.find('+') != -1:
        return int(op.split(' + ')[0]) + int(op.split(' + ')[1])
    elif op.find('/') != -1:
        return int(op.split(' / ')[0]) // int(op.split(' / ')[1])
    elif op.find('%') != -1:
        return int(op.split(' % ')[0]) % int(op.split(' % ')[1])
    elif op.find('^') != -1:
        return int(op.split(' ^ ')[0]) ** int(op.split(' ^ ')[1])
print(calc(input()))
    
```

5.8) Написать функцию, которая преобразует целое число (от 0 до 99) в текстовое написание числа на русском языке. Пример: to_text(15) -> 'Пятнадцать'

```python
def to_text(p):
    k = ''
    dic = {1: 'один',
           2: 'два',
           3: 'три',
           4: 'четыре',
           5: 'пять',
           6: 'шесть',
           7: 'семь',
           8: 'восемь',
           9: 'девять',
           10: 'десять',
           11: 'одиннадцать',
           12: 'двенадцать',
           13: 'тринадцать',
           14: 'четырнадцать',
           15: 'пятнадцать',
           16: 'шестьнадцать',
           17: 'семнадцать',
           18: 'восемнадцать',
           19: 'девятнадцать',
           20: 'двадцать',
           30: 'тридцать',
           40: 'сорок',
           50: 'пятьдесят',
           60: 'шестьдесят',
           70: 'семьдесят',
           80: 'восемьдесят',
           90: 'девяносто'}
    if 20 <= p <= 99:
        k += dic[p//10 * 10] + ' '
        p %= 10
    if 1 <= p <= 20:
        k += dic[p]
    if p == 0 and k == '':
        k = 'ноль'
    return k
print(to_text(82))
```
